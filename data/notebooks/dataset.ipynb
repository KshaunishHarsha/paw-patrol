```python
import pandas as pd
train = pd.read_csv("train.csv")
breed_labels = pd.read_csv("breed_labels.csv")
color_labels = pd.read_csv("color_labels.csv")
state_labels = pd.read_csv("state_labels.csv")
```

```python
columns = ["Breed1","Breed2","Color2","Color2","Color3","State"]
train[columns].isnull().sum()
```

```python
breed_map = dict(zip(breed_labels.BreedID, breed_labels.BreedName))
color_map = dict(zip(color_labels.ColorID, color_labels.ColorName))
state_map = dict(zip(state_labels.StateID, state_labels.StateName))
```

```python
train["Breed1"] = train["Breed1"].map(breed_map)
train["Breed2"] = train["Breed2"].map(breed_map)

train["Color1"] = train["Color1"].map(color_map)
train["Color2"] = train["Color2"].map(color_map)
train["Color3"] = train["Color3"].map(color_map)

train["State"] = train["State"].map(state_map)
```

```python
train.to_csv("pets_clean.csv", index=False)
```

```python
columns_to_keep = [
    "PetID",
    "Type",
    "Name",
    "Age",
    "Breed1",
    "Breed2",
    "Gender",
    "Color1",
    "Color2",
    "Color3",
    "MaturitySize",
    "FurLength",
    "Vaccinated",
    "Dewormed",
    "Sterilized",
    "Health",
    "Quantity",
    "Fee",
    "State",
    "Description"
]

pets = pets[columns_to_keep]
```

```python
new_columns = [
    "energy_level",
    "space_requirement",
    "grooming_level",
    "experience_required",
    "time_requirement",
    "kid_friendly",
    "pet_friendly",
    "special_needs"
]

for col in new_columns:
    pets[col] = None
```

```python
def get_energy(age):
    if age <= 12:
        return "high"
    elif age <= 60:
        return "medium"
    else:
        return "low"

pets["energy_level"] = pets["Age"].apply(get_energy)
```

```python
size_map = {
    1: "small",
    2: "medium",
    3: "large",
    4: "extra_large"
}

pets["space_requirement"] = pets["MaturitySize"].map(size_map)
```

```python
fur_map = {
    1: "low",
    2: "medium",
    3: "high"
}

pets["grooming_level"] = pets["FurLength"].map(fur_map)
```

```python
def get_experience(row):
    if row["Health"] != 1:
        return "experienced"
    elif row["Age"] < 6:
        return "intermediate"
    else:
        return "beginner"

pets["experience_required"] = pets.apply(get_experience, axis=1)
```

```python
pets["time_requirement"] = pets["energy_level"] #Pets with high enery need more attention
```

```python
def kid_friendly(age):
    if age < 3:
        return "maybe"
    else:
        return "yes"

pets["kid_friendly"] = pets["Age"].apply(kid_friendly)
```

```python
pets["pet_friendly"] = "unknown"
```

```python
pets["special_needs"] = pets["Health"].apply(lambda x: "yes" if x != 1 else "no")
```

```python
pets = pd.read_csv("pets_clean.csv")
```

```python
pets.head()
```

```python
new_columns = [
    "monthly_care_cost",
    "vet_visit_frequency",
    "medical_risk",
    "requires_temperature_control",
    "separation_anxiety_risk",
    "shedding_level",
    "expected_lifespan"
]

for col in new_columns:
    pets[col] = None
```

```python
def monthly_cost(row):

    if row["space_requirement"] in ["large", "extra_large"]:
        return "high"

    if row["grooming_level"] == "high":
        return "high"

    if row["space_requirement"] == "medium":
        return "medium"

    return "low"


pets["monthly_care_cost"] = pets.apply(monthly_cost, axis=1)
```

```python
def vet_visits(health):

    if health == 1:
        return "low"

    if health == 2:
        return "medium"

    return "high"


pets["vet_visit_frequency"] = pets["Health"].apply(vet_visits)
```

```python
def medical_risk(health):

    if health == 1:
        return "low"

    if health == 2:
        return "medium"

    return "high"


pets["medical_risk"] = pets["Health"].apply(medical_risk)
```

```python
def temperature_control(row):

    if row["FurLength"] == 3:
        return "yes"

    if row["space_requirement"] == "small":
        return "maybe"

    return "no"


pets["requires_temperature_control"] = pets.apply(temperature_control, axis=1)
```

```python
def separation_risk(age):

    if age < 12:
        return "high"

    if age < 60:
        return "medium"

    return "low"


pets["separation_anxiety_risk"] = pets["Age"].apply(separation_risk)
```

```python
fur_shedding_map = {
    1: "low",
    2: "medium",
    3: "high"
}

pets["shedding_level"] = pets["FurLength"].map(fur_shedding_map)
```

```python
def lifespan(row):

    if row["Type"] == "dog":
        return "10-13 years"

    return "12-16 years"


pets["expected_lifespan"] = pets.apply(lifespan, axis=1)
```

```python
pets.info()
```

```python
pets = pets.drop(columns=["Color2", "Color3", "Quantity", "Fee"])
```

```python
pets["Name"] = pets["Name"].fillna("Unnamed Pet")
```

```python
pets["Breed2"] = pets["Breed2"].fillna("Mixed")
```

```python
def display_breed(row):
    if row["Breed2"] == "Mixed":
        return f"{row['Breed1']} mix"
    return row["Breed1"]

pets["display_breed"] = pets.apply(display_breed, axis=1)
```

```python
pets.to_csv("pets_dataset_final.csv", index=False)
```
